```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
%matplotlib notebook

# create a new figure 
linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2
xvals = range(len(linear_data))
# plot bar chart
plt.bar(xvals, linear_data, width=0.3)
plt.title('vertical barchart')
```

```python
# add a second bar
quadratic_data = linear_data**2
new_xvals = []
for item in xvals:
    new_xvals.append(item+0.3)

plt.bar(new_xvals, quadratic_data, width=0.3, color='red')
```

```python
# add error bars using the y-error parameter
from random import randint
# create a new figure2
plt.figure()
# xxxxxxxx from previous graph
plt.bar(xvals, linear_data, width=0.3)
plt.bar(new_xvals, quadratic_data, width=0.3, color='red')
# xxxxxxxx

linear_err = [randint(0,15) for x in range(len(linear_data))]
plt.bar(xvals, linear_data, width=0.3, yerr=linear_err)
plt.title('chart with error bars')
```

```python
# stacked bar charts

plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3, color='g')
plt.bar(xvals, quadratic_data, width=0.3, bottom=linear_data, color='r')
plt.title('stacked bar chart')



```

```python
# pivot chart into a horizontal chart
plt.figure()
xvals = range(len(linear_data))
plt.barh(xvals, linear_data, height=0.3, color='g')
plt.barh(xvals, quadratic_data, height=0.3, left=linear_data, color='r')
plt.title('horizontal bar chart')

```
