## Week 3 : Custom Vizualization

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# gather neccesary data
df = pd.DataFrame(pd.read_csv('country_populations.csv'))
countries = list(df['country'])
population = (df['population']/1000000)
```

```python
# plot bar chart
plt.figure()
country = np.arange(len(countries))
bars = plt.bar(country, population, align='center', linewidth=0, color='skyblue')

def plot_chart():
    plt.xticks(country, countries)
    plt.ylabel('Population')
    plt.xlabel('Country')
    plt.title('Top 10 African Countries \nby Population on Wikipedia', alpha=0.8)
    plt.show()
```

```python
#     remove all the ticks (both axes), and tick labels on the Y axis
def clear_ticks():
    plot_chart()
    plt.tick_params(top='off', bottom='off',
                    left='off', right='off', 
                    labelleft='off', labelbottom='on')
```

```python
#     remove frame
def frameless_chart():
    clear_ticks()
    for spine in plt.gca().spines.values():
        spine.set_visible(False)
        
```

```python
# directly label the bars

def directly_labelled_chart():
    frameless_chart()          
    # direct label each bar with Y axis values
    for bar in bars:
        plt.gca().text(bar.get_x() + bar.get_width()/2,
                       bar.get_height() - 5,
                       str(int(bar.get_height())) + 'M',
                       ha='center', color='b', fontsize=9)
directly_labelled_chart()      
```

```python
# adjust the xlabels for vizibility
# aligning labels on x-axis
x = plt.gca().xaxis   
for item in x.get_ticklabels():
    item.set_rotation(45)
    # adjust for readability
plt.subplots_adjust(bottom=0.25)
```
