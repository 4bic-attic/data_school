```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
%matplotlib notebook
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2

plt.figure()
plt.plot(linear_data, '-o', quadratic_data,'-o') 
# 'o' plots point only but '-o' joins plots into a line
```

```python
# sample plot of dashed lines
plt.plot([22,44,55], '--r')
```

```python
# labeling
plt.xlabel('Linear values')
plt.ylabel('Some Y values')
# title
plt.title('Linear data  / Quadratic data')
# legend
plt.legend(['Baseline', 'Competition', 'Us'])
```

```python
# fill between function
plt.gca().fill_between(range(len(linear_data)),# get the current axes of data range
                      linear_data, quadratic_data,# lower bounds and upper bounds 
                      facecolor='green', # fill color
                      alpha=0.25) #transparency
```

```python
# create new figure
plt.figure()

observation_dates = np.arange('2017-09-01', '2017-09-09', dtype='datetime64[D]')

plt.plot(observation_dates, linear_data, '-o', 
         observation_dates, quadratic_data, '-o')
```

```python
import pandas as pd 

plt.figure()

observation_dates = np.arange('2017-09-01', '2017-09-09', dtype='datetime64[D]')

observation_dates = map(pd.to_datetime, observation_dates)
plt.plot(observation_dates, linear_data, '-o', 
         observation_dates, quadratic_data, '-o')
```
