##### Numpy Arrays | Math

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 40)
x
```

```python
sinx = np.sin(x)
sinx
```

```python
plt.plot(x,sinx)
```

```python
cosx = np.cos(x)

plt.plot(x,sinx, 'x')
plt.plot(x, cosx, 'o')

# adding a legend

plt.legend(['sin(x)', 'cos(x)'])
```
