```python
names_all = []

for year in range(1880,2014+1):
    names_all.append(pd.read_csv('names/yob{}.txt'.format(year),
                                names=['name','sex','number']))
    names_all[-1]['year'] = year
    
all_years = pd.concat(names_all)
```

```python
all_years.tail()
```

##### Names popularity

```python
allyears_indexed = all_years.set_index(['sex','name','year']).sort_index()
```

```python
allyears_indexed
```

```python
allyears_indexed.loc['F','Mary']
```

```python
def plotname(sex,name):
    data = allyears_indexed.loc[sex,name]
    
    plt.plot(data.index,data.values)
```

```python
plotname('F', 'Magdaline')
```

```python
plotname('F','Robin')
```

```python
# plot a couple of Male names

names = ['Andrew', 'Ben', 'Charles', 'Mark']

for name in names:
    plotname('M', name)
    
plt.legend(names)
```

```python
# Feminine

names = ['Sharon', 'Brenda', 'Catherine', 'Stacy']

for name in names:
    plotname('F', name)
    
plt.legend(names)
```
