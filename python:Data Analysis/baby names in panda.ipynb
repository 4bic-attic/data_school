```python
def topten(sex, year):
    simple = allyears_indexed.loc[sex,:,year].sort_values('number', ascending=False)
    simple = simple.reset_index().drop(['sex', 'year','number'],axis=1).head(10)
    
    simple.columns = [year]
    simple.index =  simple.index + 1
    
    return simple
```

```python
topten('F',1992)
```

```python
# Top ten yearly comparissons


def top_tens(sex, year0, year1):
    years = [topten(sex, year) for year in range(year0, year1+1)]
    
    return years[0].join(years[1:])
```

#### Name Fad

```python
all_years.groupby(['sex','name']).sum().head()
```

```python
totals = all_years.groupby(['sex', 'name'])['number'].sum()
```

```python
def sumsq(x):
    return sum(x**2)
```

```python
spikyness = all_years.groupby(['sex','name'])['number'].agg(sumsq) / totals**2
```

```python
spikyness.head(10)
```

```python
spiky_common = spikyness[totals > 5000]
spiky_common.sort_values(ascending=False)
spiky_common.head(10)
```

```python
spiky_common.tail(10)
```

```python
plt.figure(figsize=(12, 2.5))

plotname('F', 'Louisa')
plotname('M', 'Shaquille')

```

```python
fads = spiky_common.head(10).index.values
```

```python
plt.figure(figsize=(12, 2.5))

for sex,name in fads:
    plotname(sex,name)

plt.legend([name for sex, name in fads], loc='upper left')
```

```python
totals_recent = all_years[all_years['year'] > 2005].groupby(['sex','name'])['number'].sum()

spiky_common = spikyness[(totals > 5000) & (totals_recent < 1000)].copy()
spiky_common.sort_values(ascending=False)
spiky_common.head(10)
```

```python
fads = spiky_common.head(5).index.values
```

```python
plt.figure(figsize=(12, 2.5))

for sex,name in fads:
    plotname(sex,name)

plt.legend([name for sex, name in fads], loc='upper left')
```

### Challenge

Unisex names 

no. of times whre ratio between boys and. girls is 1:4 and 4:1

```python
totals = all_years.groupby(['sex', 'name'])['number'].sum()
totals_bysex = totals.unstack('sex')

totals_bysex.head()
```

```python
totals_both = totals_bysex.sum(axis=1)
```

```python
totals_both.head()
```

```python
ratio = totals_bysex['F'] / totals_bysex['M']
```

```python
ratio
```

```python
unisex = (ratio > 0.25) & (ratio < 4)
```

```python
totals_both = totals_both[unisex]
totals_both.sort_values(ascending=False)
totals_both.head()
```

```python
names = totals_both.head(10).index.values

plt.figure(figsize=(12,12))

for i, name in enumerate(names):
    plt.subplot(5,2,i+1)
    
    plotname('M', name)
    plotname('F', name)
    
    plt.legend([name + ' M',name + ' F'],loc='upper left')

    
```
