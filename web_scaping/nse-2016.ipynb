# Predicting stock price

```python
# Load dendencies
import csv
import numpy as np
import pandas as pd
from sklearn.svm import SVR
import matplotlib.pyplot as plt 

%matplotlib inline
# Initialize empty lists
dates = []
prices = []

# Read data from te csv file 
def get_data(filename):
    with open(filename, 'r') as csvfile:
        csvFileReader = csv.reader(csvfile)
        next(csvFileReader)
         # iterate over each row
        for row in csvFileReader:
            #use the day of the month
            dates.append(int(row[0].split('-')[0]))
            prices.append(float(row[1])) # append the opennign price
    return 

def predict_prices(dates, prices, x):
#     create a one dimensional array with numpy
    dates = np.reshape(dates,(len(dates), 1))
    #Create our model
    # Kernel > Type of SVM 
    # C > Penalty error term
    # gamma > defines how far / too far our points should exist
    
    svr_lin = SVR(kernel = 'linear', C=1e3)
    svr_poly = SVR(kernel = 'poly', C=1e3, degree =2)
    svr_rbf = SVR(kernel = 'rbf', C=1e3, gamma=0.1)
    # Fit models into training
    svr_lin.fit(dates, prices)
    svr_poly.fit(dates, prices)
    svr_rbf.fit(dates, prices)
    #plot the initial points as black and color based on model
    plt.scatter(dates, prices, color='black', label='Data')
    plt.plt(dates, svr_rbf.predict(dates), color='red',label='RBF model')
    plt.plt(dates, svr_lin.predict(dates), color='green',label='Linear model')
    plt.plt(dates, svr_poly.predict(dates), color='blue',label='Polynonial model')
    # Label X and Y axis accordingly
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Support Vector Regression')
    plt.legend()
    plt.show()
    
    # return predictions from each model
    return svr_rbf.predict(x)[0], svr_lin.predict(x)[0], svr_poly.predict(x)[0]

get_data('predict stocks/adbe.csv')

predicted_price = predict_prices(dates, prices, 10)

if __name__== main():
    print(predicted_price)

```
