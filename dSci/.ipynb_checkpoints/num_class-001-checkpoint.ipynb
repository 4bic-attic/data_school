```python
import numpy as np
import matplotlib.pyplot as plt

# Draw all plots inline in the notebook
%matplotlib inline

# Create a 1d numpy array with values 0 through 999
a = np.arange(1000)

# reshape to 10 x 10 array
a = np.reshape(a, (10, 100))

#add noise
b = np.random.normal(size=(10, 100))
a = a + b 

# inspect the first fifty Elements in the first two rows
print (a[:2, :50])

# plot the third row thro the sixth
to_plot = a[3:7, :]
for row in to_plot:
    plt.plot(row)

```
