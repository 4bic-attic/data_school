## Plot Animation

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import matplotlib.animation as animation
```

```python

n = 100
x = np.random.randn(n)
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins=bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sample the Normal Didtribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n={}'.format(curr), [3,27])
```

```python
# generate constructor
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=100)
```

# Interactivity

```python
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at Pixel{},{} and data {},{}'.format(event.x,
                                                                   event.y,
                                                                   '\n',
                                                                   event.xdata,
                                                                   event.ydata))
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

## Pick Event

```python
from random import shuffle
origins = ['Egypt', 'India', 'Kenya', 'Congo', 'South Africa',
           'Namibia', 'Ghana', 'Liberia', 'Morocco', 'Tanzania']
shuffle(origins)

df = pd.DataFrame({'height': np.random.rand(10),
                   'weight': np.random.rand(10),
                   'origin': origins })
df
```

```python
plt.figure()
plt.scatter(df['height'], df['weight'], picker=5)
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')

```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item comes from {}'.format(origin))
    
plt.gcf().canvas.mpl_connect('pick_event', onpick)
    
```
