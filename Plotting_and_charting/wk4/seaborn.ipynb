```python
%matplotlib notebook

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Generate sample data

```python
np.random.seed(1234)

# Series of normal distribution of (1000) mean(0) sd(10)
v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000), name='v2')
```

### Plot Histogram

```python
plt.figure()
plt.hist(v1, alpha=0.7, bins=np.arange(-50,150,5), label='v1');
plt.hist(v2, alpha=0.7, bins=np.arange(-50,150,5), label='v2');
plt.legend();
```

### Kernel density plot over stacked histogram

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked', normed=True);
# combine v1 & v2
v3 = np.concatenate((v1,v2))
sns.kdeplot(v3); #plot the probability density funtion of V3
```

### Combined distribution plots

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'})
```

### Joint plots

Allows for viewing of distribution of variables individualy

```python
sns.jointplot(v1, v2, alpha=0.4);
```

## Seaborn tweaks on plotting


```python
grid = sns.jointplot(v1, v2, alpha=0.4);
grid.ax_joint.set_aspect('equal')
```

#### Use Hexagon markers

```python
sns.jointplot(v1, v2,kind='hex');
```

### Heat map

```python
sns.set_style('white')

sns.jointplot(v1, v2,kind='kde', space=0);
```

### Handling of categorical data

load Iris data

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

### Scatter plot Matrix

```python
sns.pairplot(iris, hue='Name', diag_kind='kde')
```

### Violin Plot

A more informative version of box plot

```python
plt.figure(figsize=(12,8))
plt.subplot(121)
sns.swarmplot('Name', 'PetalLength', data=iris)
plt.subplot(122)
sns.violinplot('Name', 'PetalLength', data=iris)
```
