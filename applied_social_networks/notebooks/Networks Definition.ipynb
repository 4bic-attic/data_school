```python
# Multigraphs
```

```python
M=nx.MultiDiGraph()

M.add_edge('John', 'Ana', weight= 3, relation = 'siblings')
M.add_edge('Ana', 'David', weight= 4, relation = 'cousins')
M.add_edge('Ana', 'Bob', weight= 1, relation = 'friends')
M.add_edge('Ana', 'Bob', weight= 1, relation = 'neighbors')
```

```python
nx.draw_networkx(M)
```

```python
from networkx.algorithms import bipartite
```

```python
B = nx.Graph()

B.add_nodes_from(['A','B','C','D','E'], bipartite=0) #label one

B.add_nodes_from([1,2,3,4], bipartite=1) # second set of nodes1

#Add a list of edges 

B.add_edges_from([('A',1),('B',1),('C',1),('C',3),('D',2),('E',3),('E',4)])
```

```python
X = set(['A','B','C','D','E'])
```

```python
P = bipartite.projected_graph(B,X)
```

```python
nx.draw_networkx(P)
```

```python
nx.draw_networkx(B)
```
