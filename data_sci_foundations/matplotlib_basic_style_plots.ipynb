```python
%matplotlib inline 
import numpy as np
import pandas as pd 

xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

```python
df = pd.DataFrame({'x': xs, 'sinc': ys})
```

```python
df.plot.line(x='x', y='sinc')
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
df.plot.line(x='x', y='sinc')
```
