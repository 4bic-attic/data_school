```python
%matplotlib inline 
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
```

```python
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
# add a column
df['price'] = boston['target']
```

```python
hdf = df.head(10)
hdf
```

```python
# highlight max values in each column
hdf.style.highlight_max()
```

```python
# using a gradient, darker = higher values
hdf.style.background_gradient()
```

```python
# display a bar over each colums
hdf.style.bar()
```

```python
# Query cells with odd numbers
def odd_blue(val):
    if val % 2 != 0:
        return 'color: blue'
    return ''
```

```python
# color blue all odds
hdf.style.applymap(odd_blue)
```

```python
# all regions with values greater than 10, change background
def orange10(val):
    if val > 10:
        return 'background-color: orange'
    return ''
```

```python
hdf.style.applymap(orange10)
```

```python
# combination of blue values and orange background
hdf.style.applymap(odd_blue).applymap(orange10)
```

```python
# apply styles on subsets only
hdf.style.applymap(odd_blue, subset=['ZN', 'TAX'])
```

```python
# highlight top 20% values
def top20(col):
    is_top20 = col > col.quantile(0.8)
    return['font-weight: bold' if v else '' for v in is_top20]
```

```python
# top 20% values bolded
hdf.style.apply(top20)
```
