```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

```python
# simple plotting
plt.plot(xs, ys)
```

```python
# add labels and legend
plt.plot(xs, ys, label =r'$\frac{sin(x)}{x}$')
plt.legend()
```

```python
# multiple plots
plt.plot(xs, ys, label =r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.sin(xs), label ='sin(x)')
plt.legend()
```

```python
# adding subplots
plt.subplot(2,1,1)
plt.plot(xs, np.sin(xs), label ='sin')
plt.legend()

plt.subplot(2,1,2)
plt.fill(xs, -xs**2, label =r'$\{-x}^2$')
plt.legend()
# plt.plot(xs, np.cos(xs), label ='cos')
# plt.legend()
```

```python
# multiple plots
plt.plot(xs, ys, label =r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.cos(xs), label ='cos(x)')
plt.plot(xs, np.sin(xs), label ='sin(x)')
plt.legend()
```

```python
# adding subplots
plt.subplot(2,1,1)
plt.plot(xs, ys, label =r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.cos(xs), label ='cos(x)')
plt.legend()

plt.subplot(2,1,2)
plt.plot(xs, ys, label =r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.sin(xs), label ='sin(x)')
plt.legend()
```
