```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

```python
# simple plotting
plt.plot(xs, ys)
```

```python
# add labels and legend
plt.plot(xs, ys, label =r'$\frac{sin(x)}{x}$')
plt.legend()
```

```python
# multiple plots
plt.plot(xs, ys, label =r'$\frac{sin(x)}{x}$')

plt.legend()
```
